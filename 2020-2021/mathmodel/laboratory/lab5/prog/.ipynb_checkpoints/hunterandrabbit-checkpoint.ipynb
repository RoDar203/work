```python
import numpy as np
from math import sin, cos, sqrt
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
a = 0.25;
b = 0.025; 
c = 0.45; 
d = 0.045; 
```

```python
def dx(x, t):
    dx1 = - a * x[0] + c * x[0] * x[1]
    dx2 = b * x[1] - d * x[0] * x[1] 
    return [dx1, dx2]
```

```python
def draw_plot(x, y, xs, ys):
    plt.plot(x, y)
    plt.plot(xs, ys, marker='o')
    plt.title("Фазовый портрет")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()
```

```python
def draw_r_plot(x, y, t):
    plt.plot(t, x, label = 'x')
    plt.plot(t, y, label = 'x`')
    plt.title("Решение дифференциального уравнения")
    plt.xlabel('t')
    plt.ylabel('x')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
t = np.arange(0, 400, 0.1)

v0 = np.array([8, 11])

x = odeint(dx, v0, t)

xpoint = [elem[0] for elem in x] 
ypoint = [elem[1] for elem in x]

xs = a/c
ys = b/d

draw_plot(xpoint, ypoint, xs, ys)

draw_r_plot(xpoint, ypoint, t)
```
